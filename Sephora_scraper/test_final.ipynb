```python
url = "https://www.sephora.com/product/chance-eau-tendre-P258612?skuId=1237379&icid2=products%20grid:p258612"
def getPageData(url):

    page_dict = {}
    ## access url
    urlPage = requests.get(url)
    soup = bs(urlPage.text, 'html.parser')
    
    #########################################

    #brand
    try:
        brand = soup.find("a",class_="css-nc375s")
        brand = brand.text
    except:
        brand = "Not found"
    #product
    try:
        product = soup.find("span",class_="css-1pgnl76")
        product = product.text
    except:
        product = "Not found"
    
    #price
    try:
        price = soup.find("b",class_="css-0")
        price = price.text
        price = price.replace("$","").strip()
        price = float(price)
    except:
        price = "Not found"

    # stars
    try:
        stars = soup.find("div",class_="css-jp4jy6")
        stars = stars['aria-label']
        stars = stars.split("stars")[0].strip()
        stars = float(stars)
    except:
        stars = "Not found"
    
    # reviews
    try:
        reviews = soup.find("span",class_="css-1coslxg")
        reviews = reviews.text
        reviews = int(reviews)
    except:
        brand = "Not found"
   
    # size
    try:
        size = soup.find("span",class_="css-15ro776")
        size = size.text
    except:
        size = "Not found"
    
    # love
    try:
        love = soup.find("span",class_="css-jk94q9")
        love = love.text
    except:
        love = "Not found"
    
    page_dict['brand'] = brand
    page_dict['product'] = product
    page_dict['price'] = price
    page_dict['stars'] = stars
    page_dict['reviews'] = reviews
    page_dict['size'] = size
    page_dict['love'] = love

    return page_dict

print(getPageData(url))
```

```python


def crawl_urls():
    urls = []
    homePage = requests.get("https://www.sephora.com/shop/fragrances-for-women")
    soup = bs(homePage.content, 'html.parser')
    totalPages = 19

    for pageNum in range(totalPages):
        pageUrl = "https://www.sephora.com/shop/fragrances-for-women?currentPage={0}".format  (pageNum)
        page = requests.get(pageUrl)
        soup = bs(page.content, 'html.parser')
        
        product_link = soup.find_all("a", attrs = {"class": "css-ix8km1"})
        
        for link in product_link:
            product_href = 'https://sephora.com' + link.get('href')
            urls.append(product_href)
            print('crawling equipment page: ', product_href)
        
        print('done crawling page: ', pageUrl)
    return urls

urls = crawl_urls()
```

```python
def getAllData(urls):
    data_list = []
    for url_index in range(len(urls)):
        data_list.append(getPageData(urls[url_index]))
        print("Finished...")
    data = pd.DataFrame(data_list)
    print("Done.")
    return data
```

```python
data = getAllData(urls)
def save(df):
    df.to_csv("example.csv",index = False, sep=",")
    print("Successfully transformed and saved to csv.")
save(data)
```
